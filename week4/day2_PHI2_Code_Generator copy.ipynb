```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_name = "microsoft/phi-2"  # Much faster and works on CPU

# Load tokenizer and quantized model (auto uses GPU if available)
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
```

```python
# ✅ Load full precision model for CPU
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float32,       # ✅ Best for CPU stability
    device_map="cpu",                # ✅ Force CPU-only
    trust_remote_code=True
)
```

```python
# Prompt setup
prompt = "write a quick sort JS algorithm."
messages = [
    {"role": "system", "content": "You are Qwen, created by Alibaba Cloud. You are a helpful assistant."},
    {"role": "user", "content": prompt}
]
# text = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
# model_inputs = tokenizer([text], return_tensors="pt").to(model.device)
model_inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
```

```python
# Fast generation => 64 for Testing Purpuse
with torch.inference_mode():
    generated_ids = model.generate(
        **model_inputs,
        max_new_tokens=128,        # Minimized token count
        do_sample=False,          # Greedy for speed
        use_cache=True,
        early_stopping=True       # ⬅️ stop when appropriate
    )
```

```python
# Remove prompt from output and decode
generated_ids = [
    output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
]
```

```python
response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
print(response)
```
