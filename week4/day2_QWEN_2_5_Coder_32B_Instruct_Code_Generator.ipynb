```python
!pip install -q bitsandbytes
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
import torch

# ✅ 模型名稱（需具備 64GB+ VRAM）
model_name = "Qwen/Qwen2.5-Coder-32B-Instruct"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    llm_int8_enable_fp32_cpu_offload=True
)

# ✅ 載入 tokenizer 和模型，使用 float16 + 自動分配設備
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
```

```python
# Take longer time, please wait ...
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map="auto",
    trust_remote_code=True,
    # offload_buffers=True
)
```

```python
# ✅ Prompt（輸入任務）
prompt = "write a quick sort JS algorithm."
messages = [
    {"role": "system", "content": "You are Qwen, created by Alibaba Cloud. You are a helpful assistant."},
    {"role": "user", "content": prompt}
]

# ✅ 將聊天格式轉換為文字輸入
text = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
inputs = tokenizer(text, return_tensors="pt").to(model.device)
```

```python
# ✅ 使用自動混合精度進行推理（新版語法，無警告）
with torch.inference_mode():
    output_ids = model.generate(
      **inputs,
      max_new_tokens=256,
      do_sample=False,  # 保留這行
      eos_token_id=tokenizer.eos_token_id
      # 不要設置 temperature, top_p, top_k
    )
```

```python
# ✅ 擷取新生成的 token（排除 prompt）
generated_ids = [
    output_ids[i][len(inputs.input_ids[i]):]
    for i in range(len(output_ids))
]
```

```python
# ✅ 解碼為可讀文字
response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]

# ✅ 顯示回應
print(response)
```
